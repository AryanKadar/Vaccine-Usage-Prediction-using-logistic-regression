```python
import pandas as pd
import matplotlib.pyplot as mp
from sklearn.preprocessing import OneHotEncoder as OHE
from sklearn.metrics import confusion_matrix as CM,precision_score as PS,recall_score as RS
from sklearn.compose import ColumnTransformer as CT
from sklearn.pipeline import Pipeline as PP
from sklearn.linear_model import LogisticRegression as LOR
from sklearn.preprocessing import StandardScaler as SS
from sklearn.model_selection import train_test_split as TTS,GridSearchCV as GSCV,cross_val_score as CVS
from sklearn import set_config
set_config(display="diagram")
```

```python
t=pd.read_csv("h1n1_vaccine_prediction.csv")
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
t.head()
```

```python
t.drop(["unique_id",'race',"income_level",'housing_status','employment','qualification','census_msa',"age_bracket",'no_of_adults','no_of_children'],axis=1,inplace=True)
```

```python
t.head()
```

```python
t.isnull().sum()/len(t)*100
```

```python
fig,ax=mp.subplots(figsize=(15,8))
t.isnull().sum().plot(kind="bar",rot=70);
```

```python
t.drop("has_health_insur",axis=1,inplace=True)
```

```python
t.dropna(inplace=True)
```

```python
t.isnull().sum()
```

```python
t.tail()
```

```python
ct=CT([
    ("OHE",OHE(drop="first"),['sex','marital_status'])
],remainder="passthrough")
```

```python
x=t.iloc[:,:-1]
x.head()
```

```python
y=t.iloc[:,-1]
y.head()
```

```python
xta,xte,yta,yte=TTS(x,y,train_size=0.2,random_state=4)
```

```python
pp=PP(
[
    ("CT",ct),
    ("SS",SS()),
    ("logr",LOR())
]
)
pp.fit(xta,yta)
```

```python
pp.named_steps["logr"].intercept_
```

```python
pp.named_steps["logr"].coef_
```

```python
CVS(pp,xta,yta,cv=5,scoring="accuracy").mean()
```

```python
p={
    "logr__penalty":['l1', 'l2'],
    "logr__solver":['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
    ,"logr__max_iter":[1,25,50,100,150,200,500]
}
```

```python
g=GSCV(pp,param_grid=p,cv=5,n_jobs=-1,scoring="accuracy")
```

```python
g.fit(xta,yta)
```

```python
g.best_score_
```

```python
g.best_params_
```

```python
ypre=g.predict(xte)
ypre
```

```python
CM(yte,ypre)
```

```python
PS(yte,ypre)
```

```python
RS(yte,ypre)
```
